## Punto 3
Tomar las imágenes del directorio dataset/raw y unificar la resolución de las imágenes para que sea 800x600. Las nuevas imágenes deben quedar almacenadar en dataset/prc/resize. Solo tomar las imágenes válidas. El directorio prc/resize debe ser traceado por dvc.

Escribir los pasos necesarios para el traceo del dataset y hacer los commit correspondientes y pushear dvc.

```python
import os
import cv2

# Rutas de directorio de origen y destino
directorio_origen = "/content/Chessman-image-dataset-Bishop/Chess/Bishop"
directorio_destino = "/dataset/prc/resize"

# Crear el directorio de destino si no existe
os.makedirs(directorio_destino, exist_ok=True)

# Lista de extensiones de archivo válidas (puedes personalizar esta lista)
extensiones_validas = ['.jpg', '.jpeg', '.png', '.JPG']

for filename in os.listdir(directorio_origen):
    if any(filename.endswith(ext) for ext in extensiones_validas):
        # Cargar la imagen desde el directorio de origen
        imagen = cv2.imread(os.path.join(directorio_origen, filename))

        # Verificar si la imagen se cargó correctamente
        if imagen is not None:
            # Redimensionar la imagen a 800x600 píxeles
            imagen_redimensionada = cv2.resize(imagen, (800, 600))

            # Guardar la imagen redimensionada en el directorio de destino
            cv2.imwrite(os.path.join(directorio_destino, filename), imagen_redimensionada)

print("Proceso de redimensionamiento completado.")
```

Tomar las imágenes del directorio dataset/prc/resize y pasar las imágenes a blanco y negro. Las nuevas imágenes deben quedar almacenadar en dataset/prc/black_white. El directorio dataset/prc/black_white debe ser traceado por dvc.

Escribir los pasos necesarios para el traceo del dataset y hacer los commit correspondientes y pushear dvc.

```python
import os
import cv2

# Rutas de directorio de origen (imágenes a blanco y negro) y destino
directorio_origen = "/dataset/prc/resize"
directorio_destino = "/dataset/prc/black_white"

# Crear el directorio de destino si no existe
os.makedirs(directorio_destino, exist_ok=True)

for filename in os.listdir(directorio_origen):
    ruta_origen = os.path.join(directorio_origen, filename)
    ruta_destino = os.path.join(directorio_destino, filename)

    # Cargar la imagen desde el directorio de origen
    imagen = cv2.imread(ruta_origen)

    # Verificar si la imagen se cargó correctamente
    if imagen is not None:
        threshold = 127
        maxval = 255
        # Convertir la imagen a blanco y negro
        imagen_bn = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
        _, binarizada = cv2.threshold(imagen_bn, 210, 255, cv2.THRESH_BINARY)
        # Guardar la imagen en blanco y negro en el directorio de destino
        cv2.imwrite(ruta_destino, binarizada)

print("Proceso de conversión a blanco y negro completado.")


```

### Punto 5
Tomar las imágenes del directorio dataset/prc/resize y recotar la imágen donde solo quede centrada la pieza de ajedrez. Las nuevas imágenes deben quedar almacenadar en dataset/selection.

Este recorte debe ser una aproximación, no va a ser exacto porque existen imágenes complejas.

Solo debo guardar las imágenes que admiten algún tipo de recorte, ya sea horizontal o vertical.

Escribir los pasos necesarios para el traceo del dataset y hacer los commit correspondientes y pushear dvc.

```python
import os
import cv2

# Rutas de directorio de origen (imágenes a recortar) y destino
directorio_origen = "/dataset/prc/resize"
directorio_destino = "/dataset/selection"

# Crear el directorio de destino si no existe
os.makedirs(directorio_destino, exist_ok=True)

# Definir un umbral para determinar si la imagen es candidata a recorte
umbral = 0.6  # Puedes ajustar este valor según tus necesidades

for filename in os.listdir(directorio_origen):
    ruta_origen = os.path.join(directorio_origen, filename)
    ruta_destino = os.path.join(directorio_destino, filename)

    # Cargar la imagen desde el directorio de origen
    imagen = cv2.imread(ruta_origen)

    # Verificar si la imagen se cargó correctamente
    if imagen is not None:
        # Convertir la imagen a escala de grises para facilitar el análisis
        imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

        # Calcular el histograma de la imagen
        histograma = cv2.calcHist([imagen_gris], [0], None, [256], [0, 256])

        # Calcular la proporción de píxeles blancos en la imagen
        proporcion_blancos = histograma[255][0] / imagen_gris.size

        # Verificar si la imagen es candidata a recorte (proporción alta de píxeles blancos)
        if proporcion_blancos >= umbral:
            # Realizar el recorte centrado (ajuste la región de interés según tus necesidades)
            altura, ancho = imagen.shape[:2]
            margen = 50  # Margen de recorte
            roi = imagen[margen:altura - margen, margen:ancho - margen]

            # Guardar la imagen recortada en el directorio de destino
            cv2.imwrite(ruta_destino, roi)

print("Proceso de recorte completado.")
```
