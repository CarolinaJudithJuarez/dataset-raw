# Ejercicio 04

Dado un arreglo se debe crear una martiz NxM y otra matriz MxL que sea inversa a la anterior y mostrar el resultado de la multiplicación de ambas matrices.

Ejemplo de matriz inversa para el ejercicio:

matriz 1 = [[1,2], [3,4]]

matriz 2 = [[4,3], [2,1]] # matriz 1 inversa

```python
arreglo = np.linspace(0, 12, 24)

#
arreglo_2 = arreglo[::-1]
# matriz 1 de 4x6
matriz_1 = arreglo.reshape(4,6)
# matriz 2 de 6x4
matriz_2 = arreglo_2.reshape(6,4)
# agrego las 2 primeras filas a la matriz
matriz_1 = np.concatenate([matriz_1, np.vsplit(matriz_1,[2])[0]], axis=0)
# agrego las 2 primeras columnas a la matriz
matriz_2 = np.concatenate([matriz_2, np.hsplit(matriz_2,[2])[0]], axis=1)

print(f"El resultado de la multiplicación es: \n", np.multiply(matriz_1, matriz_2))
```

# Ejercicio 05

Dado el siguiente arreglo, generar una matrizde 10 filas y 20 columnas para luego normalizar los valores de la matriz.

Mostrar ambas matrices.

Para normalizar una matriz se debe tomar el mayor valor de la matriz y dividirlo por cada valor de la matriz, esto hace que todos los valores queden entre 0 y 1.

```python
arreglo = np.random.randint(low=0, high=12345, size=200)
###
matriz_normal = arreglo.reshape(10,20)
maximo = arreglo.max()
matriz_normalizada = np.divide(matriz_normal, maximo)
print("La matriz original es:\n", matriz_normal)
print()
print("La matriz normalizada es:\n", matriz_normalizada)
```

# Ejercicio 06

Dada una matriz obtener todos los valores mayores a la mediana y generar una nueva matriz donde la posición es 0 si el valor es menor o igual a la mediana y en caso contrario el valor.

Mostrar ambas matrices y el valor de la mediana.

```python
matriz = np.random.randint(low=0, high=12345, size=200).reshape(10,20)
###
mediana = np.nanmedian(matriz)

print("La matriz original es:\n", matriz)
print("La mediana es:\n", mediana)
print("La nueva matriz es:\n", np.where(np.greater(matriz, mediana), matriz, 0))
```

# Ejercicio 07

Crear un gráfico del logaritmo de 100 puntos equidistantes entre 0.000001 y 100.

El gráfico debe tener 100 dpi, la línea debe ser de color purpura, la línea debe ser con guiones. Debe poseer título, leyendas y grilla.

```python
valores_x = np.linspace(0.00001, 100, 100)
valores_y = np.log(valores_x)
figura = plt.figure(dpi=100)
plt.plot(valores_x, valores_y, linestyle="--", color="purple")
plt.title("Logaritmo")
plt.xlabel("X")
plt.ylabel("log(x)")
plt.legend()
plt.grid()
```
